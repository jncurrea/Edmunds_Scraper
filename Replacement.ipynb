```python

# Filepaths
output_file = 'replacement_sample_data.csv' # store modified data 
input_file = 'edmunds_scraper.csv' # original data
replacement_file = 'Replacement_words.csv' # original and replacement words

# temporary file to wrtie the changes before moving it to the final location
tempfile = NamedTemporaryFile(mode='w', delete=False, newline='', encoding='utf-8')


def load_replacements(replacement_file):

    replacements = {}
    with open(replacement_file, 'r', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile, delimiter=',', quotechar='"')
        next(reader)  # skip first row with Brand and Model
        for row in reader:
            brand, model = row[0].strip().lower(), row[1].strip().lower()  # Ensure lowercase comparison
            replacements[model] = brand
    return replacements

def replace_words_in_text(text, replacements):

    text = text.lower()  # Convert the entire text to lowercase first
    for model, brand in replacements.items():
        pattern = re.compile(r'\b' + re.escape(model) + r'\b', re.IGNORECASE)  # Match whole words
        text = pattern.sub(brand, text)
    return text

def process_file(input_file, output_file, replacements):

    with open(input_file, 'r', encoding='utf-8') as csvfile, \
         open(output_file, 'w', newline='', encoding='utf-8') as temp_output:
        reader = csv.reader(csvfile)
        writer = csv.writer(temp_output)
        
        # Process each row
        for row in reader:
            modified_row = [replace_words_in_text(cell, replacements) for cell in row]
            writer.writerow(modified_row)


# Load the replacement words from the replacement CSV file
replacements = load_replacements(replacement_file)

# Process the input file and apply the replacements
process_file(input_file, output_file, replacements)


```
